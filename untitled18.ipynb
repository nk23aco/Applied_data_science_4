```python
def plot_active_inactive_members(data):
    """
    Plot the number of people active and inactive based on gender (Male or Female).

    Arguments:
        data (DataFrame): 'Gender' and 'IsActiveMember' columns in a pandas DataFrame.

    Returns:
        None
    """
    is_active_member_male = 0
    is_active_member_female = 0
    inactive_member_male = 0
    inactive_member_female = 0

    for index, row in data.iterrows():
        # Handling unexpected values in 'Gender' or 'IsActiveMember' columns
        gender = row['Gender']
        is_active = row['IsActiveMember']
        if gender not in ['Male', 'Female']:
            raise ValueError(f"Unexpected value '{gender}' in 'Gender' column.")
        if is_active not in [0, 1]:
            raise ValueError(f"Unexpected value '{is_active}' in 'IsActiveMember' column.")

        if gender == 'Male' and is_active == 1:
            is_active_member_male += 1
        elif gender == 'Male' and is_active == 0:
            inactive_member_male += 1
        elif gender == 'Female' and is_active == 1:
            is_active_member_female += 1
        else:
            inactive_member_female += 1

    # Labels for the bar graph.
    labels = ['Active Male (1)', 'Active Female (1)', 'Inactive Male (0)', 'Inactive Female (0)']
    counts = [is_active_member_male, is_active_member_female, inactive_member_male, inactive_member_female]
    # Size of the graph 8 X 6.
    plt.figure(figsize=(8, 6))
    # Plotting the bar graph and passing the parameters in function.
    bars = plt.bar(labels, counts, color=['blue', 'blue', 'pink', 'pink'])

    # Add annotations to each bar.
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, height, ha='center', va='bottom')
    # Adding the labels and title for the graph.
    plt.xlabel('Gender and Activity', fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.title('Counts of Active and Inactive Members by Gender', fontsize=14)
    plt.show()

# Calling the function.
plot_active_inactive_members(data)
```

```python
def pie_chart(data):
    """
    Ploting  the share of males and females from the given DataFrame.

    Parameters:
        data: A pandas DataFrame containing a 'Gender' column with values 'Male' or 'Female'.

    """
    # Initialize counts for males and females
    male_count = 0
    female_count = 0

    # Iterate through each row in the DataFrame
    for index, row in data.iterrows():
        # checking the row with gender
        # and update the count
        gender = row['Gender']
        if gender == 'Male':
            male_count += 1
        elif gender == 'Female':
            female_count += 1
        else:
            # handling value error, if any exception occurs.
            raise ValueError(f"Unexpected value '{gender}' in 'Gender' column.")

    # Labels for a pie chart
    labels = ['Male', 'Female']
    # shares for the chart
    sizes = [male_count, female_count]
    # colors for the each share for the chart
    colors = ['blue', 'pink']
    # creating a pie chart with angle 140
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    # axis in equal.
    plt.axis('equal')
    # adding the title for the graph.
    plt.title('Distribution of Males and Females', fontsize=14)
    plt.show()

# Calling the function.
pie_chart(data)
```

```python
def scatter_plot(data):
    """
    Plotting a scatter plot, age against NumOfProducts.

    Parameters:
        data (DataFrame): DataFrame containing the 'Age' and 'NumOfProducts' columns.

    Returns:
        None
    """
    # Plotting the scatter plot between age and number of products.
    plt.scatter(data['Age'], data['NumOfProducts'], alpha=0.7)

    # Adding x and ylabels and title for the scatter plot.
    plt.xlabel('Age', fontsize=14)
    plt.ylabel('NumOfProducts', fontsize=14)
    plt.title('Age vs. NumOfProducts', fontsize=14)

    # Display the plot.
    plt.show()

# Calling the function.
scatter_plot(data)
```

```python
def violin_plot(data):
    """
    Plotting a 'Age' by 'Exited' to visualise in violin plot of these features..

    Parameters:
        data : A pandas DataFrame containing 'Age' and 'Exited' columns.

    """
    # Size of the plot 10 X 6.
    plt.figure(figsize=(10, 6))
    # Plotting the violin plot between age and exited.
    sns.violinplot(x='Exited', y='Age', data=data, palette='Set1', hue='Exited')
    # Adding the title and labels for the X and Y-axis.
    plt.title('Distribution of Age by Exited', fontsize=14)
    plt.xlabel('Exited', fontsize=14)
    plt.ylabel('Age', fontsize=14)
    # Adding the xticks for the X labels.
    plt.xticks(ticks=[0, 1], labels=['Not Exited', 'Exited'], fontsize=14)
    plt.show()

# Calling the function.
violin_plot(data)
```

```python
def line_fitting(x, y):
    """
    Perform linear fitting and visualize the fitted line with confidence interval.

    Parameters:
        x (credit score): Independent variable (x-axis).
        y (Balance): Dependent variable (y-axis).

    Returns:
        tuple: Tuple containing slope, intercept, and confidence interval of the fitted line.
    """
    # Fiting a degree 1 using polyfit function.
    coefficients = np.polyfit(x, y, 1)

    # getting slope and y-intercept from the coefficients
    m, b = coefficients

    # Generate predicted y-values based on the line equation.
    y_pred = m * x + b

    # Calculating residuals
    residuals = y - y_pred

    # calculating the variance for the residuals
    variance = np.sum(residuals ** 2) / (len(x) - 2)

    # Calculating the standard error of the estimate
    std_error = np.sqrt(variance)

    # Calculate the t-value for a 95% confidence interval.
    t_value = 1.96

    # Calculating the confidence interval.
    confidence_interval = t_value * std_error

    # Plot the figure with size 10 X 6.
    plt.figure(figsize=(10, 6))
    # plotting the data points on the axis using scatter plot.
    plt.scatter(x, y, color='blue', label='Data points')
    # Fitting the best fit line on the data points.
    plt.plot(x, y_pred, color='red', label='Fitted line')

    # Plot error bars for the confidence interval
    plt.fill_between(x, y_pred - confidence_interval, y_pred + confidence_interval, color='gray', alpha=0.2)
    #  Adding the X and Y labels for the plot.
    plt.xlabel('CreditScore', fontsize=14)
    plt.ylabel('Balance', fontsize=14)
    # add the titles for the plot.
    plt.title('Linear Fitting with Confidence Interval (CreditScore vs. Balance)', fontsize=14)
    plt.legend()
    # Adding the grid.
    plt.grid(True)
    plt.show()

    return m, b, confidence_interval

# Independent variable (x-axis)
x = data['CreditScore'].values
# Dependent variable (y-axis)
y = data['Balance'].values

# fitting the best fit line along with confidence interval 95%.
slope, intercept, confidence_interval = line_fitting(x, y)
# printing the slope.
print("Slope:", slope)
# printing the intercept.
print("Intercept:", intercept)
# printing the value at the confidence interval 95%.
print("Confidence Interval (95%):", confidence_interval)
```

```python
# global variable
scaler = StandardScaler()

def preprocess_data(data):
    """
    Standardizing the 'Age' and 'Balance' values in a columns .

    Parameters:
        data: DataFrame containing 'Age' and 'Balance' columns.

    Returns:
        ndarray: Standardized data.
    """
    return scaler.fit_transform(data[['Age', 'Balance']])

def perform_elbow_method(data_scaled):
    """
    calculating the optimal value using elbow method for clusters.

    Parameters:
        data_scaled: Standardized data.

    Returns:
        integer: Optimal number of clusters.
    """
    wcss = []
    for i in range(1, 11):
        kmeans = KMeans(n_clusters=i, random_state=0)
        kmeans.fit(data_scaled)
        wcss.append(kmeans.inertia_)
    plt.plot(range(1, 11), wcss, marker='o')
    plt.xlabel('Number of Clusters')
    plt.ylabel('WCSS')
    plt.title('Elbow Method')
    plt.grid(True)
    plt.show()
    return 4

def visualize_clusters(data_scaled, optimal_num_clusters):
    """
    Visualizing the clusters.

    Parameters:
        data_scaled (ndarray): Standardized data.
        optimal_num_clusters (int): Optimal number of clusters.
    """
    kmeans = KMeans(n_clusters=optimal_num_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(data_scaled)
    cluster_centers = kmeans.cluster_centers_

    plt.figure(figsize=(8, 6))
    for cluster_label in range(optimal_num_clusters):
        plt.scatter(data_scaled[cluster_labels == cluster_label, 0],
                    data_scaled[cluster_labels == cluster_label, 1],
                    label=f'Cluster {cluster_label + 1}')

    plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', color='black', label='Cluster Centers')

    plt.xlabel('Standardized Age')
    plt.ylabel('Standardized Balance')
    plt.title('KMeans Clustering')
    plt.legend()
    plt.grid(True)
    plt.show()

def predict_new_data(X, kmeans):
    """
    Predicts cluster labels for new data points.

    Parameters:
        X (ndarray): New data points.
        kmeans (KMeans): Trained KMeans model.

    Returns:
        ndarray: Predicted cluster labels for new data points.
    """
    scaled_X = scaler.transform(X)
    return kmeans.predict(scaled_X)

# preprocessing the data
data_scaled = preprocess_data(data)

# calulating the optimal value for determing the clusters using elbow method.
optimal_num_clusters = perform_elbow_method(data_scaled)

# Performing KMeans clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_num_clusters, random_state=0)
cluster_labels = kmeans.fit_predict(data_scaled)

# Visualizing the clusters
visualize_clusters(data_scaled, optimal_num_clusters)
```

```python
# sample data for prediction
X_new = np.array([[30, 50000], [40, 60000], [50, 70000], [90, 100000]])
predicted_labels = kmeans.predict(scaler.transform(X_new))
# Generate colors randomly based on the number of clusters
colors = plt.cm.viridis(np.linspace(0, 1, optimal_num_clusters))

# Plot the new data points with colors corresponding to their predicted clusters
for i, label in enumerate(predicted_labels):
    plt.scatter(X_new[i, 0], X_new[i, 1], color=colors[label], label=f'Predicted Cluster {label}')
plt.xlabel('Standardized Age', fontsize = 14)
plt.ylabel('Standardized Balance', fontsize = 14)
plt.title('KMeans Clustering', fontsize = 14)
plt.legend()
plt.grid(True)
plt.show()
# Predict cluster labels for new data points
predicted_labels = predict_new_data(X_new, kmeans)
print("Predicted labels for new data points:", predicted_labels)
```
